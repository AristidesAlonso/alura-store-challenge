# 3. Calificación promedio de la tienda


```python
import matplotlib.pyplot as plt
import numpy as np

# Calcular calificación promedio de cada tienda
promedio1 = tienda['Calificación'].mean()
promedio2 = tienda2['Calificación'].mean()
promedio3 = tienda3['Calificación'].mean()
promedio4 = tienda4['Calificación'].mean()

# Mostrar resultados numéricos
print("CALIFICACIÓN PROMEDIO POR TIENDA")
print("-" * 35)
print(f"Tienda 1: {promedio1:.2f}")
print(f"Tienda 2: {promedio2:.2f}")
print(f"Tienda 3: {promedio3:.2f}")
print(f"Tienda 4: {promedio4:.2f}")
print("\n")

# Datos para gráfico
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios = [promedio1, promedio2, promedio3, promedio4]
colores = ['#4CAF50', '#2196F3', '#FF9800', '#F44336']

# Gráfico de barras (mejor opción para comparar promedios)
plt.figure(figsize=(10, 6))
barras = plt.bar(tiendas, promedios, color=colores, edgecolor='black', linewidth=1.2)

# Agregar valores sobre las barras
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2., altura + 0.05,
             f'{altura:.2f}',
             ha='center', va='bottom', fontsize=12, fontweight='bold')

# Línea de referencia para calificación máxima (5)
plt.axhline(y=5, color='gray', linestyle='--', alpha=0.7, label='Máxima (5)')

# Personalización
plt.title('Calificación Promedio por Tienda', fontsize=16, fontweight='bold')
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.ylim(0, 5.5)  # Escala de 0 a 5
plt.grid(axis='y', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Análisis de Calificaciones Promedio por Tienda

| Tienda 3 | 4.05 |
| Tienda 4 | 4.00 |

### Análisis narrativo
Los datos muestran una **satisfacción general uniforme** en todas las tiendas, con calificaciones cercanas a 4 sobre 5:

- **Tienda 3** tiene la mejor calificación con **4.05**, siendo la mejor evaluada por los clientes
- **Tienda 2** le sigue muy de cerca con **4.04**
- **Tienda 4** obtiene exactamente **4.00**, en el promedio general
- **Tienda 1** presenta la calificación más baja con **3.98**, aunque la diferencia es mínima

### Observaciones clave
- Las 4 tiendas están en un rango muy estrecho de **solo 0.07 puntos** de diferencia
- Ninguna tienda alcanza el 4.5, pero todas superan el 4 (excepto Tienda 1 por 0.02)

### Conclusión parcial
En términos de satisfacción del cliente, **no hay una tienda claramente deficiente**. La Tienda 1 tiene la calificación más baja pero por un margen insignificante (0.02). Este criterio no sería determinante para decidir qué tienda vender.
